```python
#抓取學校地址經緯度
import requests
import json
import time

file_name = "./school/taichung_school.csv"
file_object = open(file_name,"r")
data = np.array([row for row in csv.reader(file_object)])
attr = data[0][0:5]+['location']
f = open("./school/taichung_school_loc.csv","w")
f.write(",".join(attr))
f.write("\n")
for row in data[1:]:
    addr = row[4]
    res = requests.get('https://maps.googleapis.com/maps/api/geocode/json?address='+addr+'&key=AIzaSyBCDNF8aSNQR6STp814uuav8o1mS9Sf4Nk')
    jd = json.loads(res.text)
    for post in jd['results']:
        location = post['geometry']['location']['lat'],post['geometry']['location']['lng']
        f.write(",".join(row[0:5]+[str(location)]))
        f.write("\n")
        time.sleep(1.5)
        break
        
f.close()        

```

```python
#去除經緯度逗號及括號
file_name = "./school/taichung_school_loc.csv"
file_object = open(file_name,"r")
data = np.array([row for row in csv.reader(file_object)])

f = open("./school/taichung_school_loc1.csv","w")
f.write(",".join(data[0]))
f.write("\n")
for row in data[1:]:
    loc = row[5].strip("(")+row[6].strip("(").strip(")")
    f.write(",".join(row[:5]+[loc]))
    f.write("\n")
f.close()     
```

```python
#將各地區學校各自存成CSV檔
import os, sys

file_name = "./school/taichung_school_loc1.csv"
file_object = open(file_name,"r")
data = np.array([row for row in csv.reader(file_object)])

for row in data[1:]:
    loc_num = row[4][1:4]
    f = open("./school/town/"+loc_num+".csv","a")
    f.write(",".join(row))
    f.write("\n")
    f.close()
    
# Open a file
path = "./school/town/"
dirs = os.listdir( path )

# This would print all the files and directories
for file in dirs:
    with open(path+file,newline='') as f:
        r = csv.reader(f)
        data1 = [line for line in r]
    with open(path+file,'w',newline='') as f:
        w = csv.writer(f)
        w.writerow(data[0])
        w.writerows(data1)
```

```python
#統計各地區各級學校數量
import os, sys

path = "./school/town/"
dirs = os.listdir( path )

# This would print all the files and directories
for file in dirs:
    cnt1 = 0 #國小
    cnt2 = 0 #國中
    cnt3 = 0 #高中
    cnt4 = 0 #高職
    cnt5 = 0 #專科
    cnt6 = 0 #大學
    cnt7 = 0 #大學進修
    cnt8 = 0 #特殊
    cnt9 = 0 #其他
    file_name = path + file
    file_object = open(file_name,"r")
    data = np.array([row for row in csv.reader(file_object)])
    for row in data[1:]:
        school_class_id = row[0][-3]
        school_class_name = row[1][-2:]
        if school_class_id == '6' or school_class_id == '7' or school_class_id == '8':
            cnt1 = cnt1 + 1
        elif school_class_id == '5':
            cnt2 = cnt2 + 1
        elif school_class_id == '3':
            if school_class_name == '國小':
                cnt1 = cnt1 + 1
            elif school_class_name == '國中':
                cnt2 = cnt2 + 1
            else:
                cnt3 = cnt3 + 1
        elif school_class_id == '4':
            if school_class_name == '國小':
                cnt1 = cnt1 + 1
            elif school_class_name == '國中':
                cnt2 = cnt2 + 1
            else:
                cnt4 = cnt4 + 1
        elif school_class_id == '2':
            cnt5 = cnt5 + 1
        elif school_class_id == '1' or school_class_id == '0':
            cnt6 = cnt6 + 1
        elif school_class_id == 'A':
            cnt7 = cnt7 + 1
        elif school_class_id == 'F':
            cnt8 = cnt8 + 1
        else:
            cnt9 = cnt9 + 1
    f = open("./school/taichung_school_num.csv","a")
    result = [file[:3],str(cnt1),str(cnt2),str(cnt3),str(cnt4),str(cnt5),str(cnt6),str(cnt7),str(cnt8),str(cnt9)]
    f.write(",".join(result))
    f.write("\n")
    f.close()

with open('./school/taichung_school_num.csv',newline='') as f:
    r = csv.reader(f)
    data1 = [line for line in r]
with open('./school/taichung_school_num.csv','w',newline='') as f:
    w = csv.writer(f)
    w.writerow(['郵遞區號','國小數量','國中數量','高中數量','高職數量','專科學校數量','大專院校數量','大專進修學校數量','特殊學校數量','其他'])
    w.writerows(data1)
```
